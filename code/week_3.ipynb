```python
import numpy as np
import seaborn as sns

x=np.random.normal(0,1,(10000,1))
x1=np.random.normal(5,1,(100,1))
x=np.concatenate((x,x1))
```

```python
sns.histplot(x)
```

```python
data_mean=np.mean(x)
print(data_mean)
data_std=np.std(x)
print(data_std)
data_max=data_mean+data_std*3
print(data_max)
data_min=data_mean-data_std*3
print(data_min)

x_new=x[(x<data_max)&(x>data_min)]
```

```python
sns.histplot(x_new)
```

```python

sns.boxenplot(x,palette='cool')
```

```python
x_sorted=np.sort(x[:,0])
q1=x_sorted[int(len(x_sorted)/4)]
q2=x_sorted[int(len(x_sorted)/2)]
q3=x_sorted[3*int(len(x_sorted)/4)]
iqr = q3-q1
minn = q1-1.5*iqr
maxx = q3+1.5*iqr
print(minn)
print(maxx)

x_iqr=x[(x<maxx)&(x>minn)]
```

```python
print(q1)
print(q2)
print(q3)
```

```python
from sklearn.cluster import DBSCAN

random_data = np.random.randn(10000,2)  * 10 + 20
random_data1 = np.random.randn(100,2)  * 5 + 70
random_data=np.concatenate((random_data,random_data1))
```

```python
random_data.shape
```

```python
sns.scatterplot(x=random_data[:,0],y=random_data[:,1])
```

```python
outlier_detection = DBSCAN(min_samples = 10, eps = 20)
clusters = outlier_detection.fit_predict(random_data)
```

```python
sns.scatterplot(x=random_data[:,0],y=random_data[:,1],hue=clusters)
```

```python
from sklearn.ensemble import IsolationForest

clf = IsolationForest(random_state=0,n_estimators=100,verbose=1)
preds = clf.fit_predict(random_data)
```

```python
preds
```

```python
sns.scatterplot(x=random_data[:,0],y=random_data[:,1],hue=preds)
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)

sns.set_theme()
sinplot()
```
